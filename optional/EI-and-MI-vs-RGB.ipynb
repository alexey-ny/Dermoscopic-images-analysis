```python
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
from mpl_toolkits.mplot3d import axes3d

import warnings
warnings.filterwarnings('ignore')
import cv2
import numpy as np
import pandas as pd 
import math
```

```python
red = np.zeros(255)
green = np.zeros(255)
mi = np.zeros(255)
mi_norm = np.zeros(255)
```

```python
for i in range(255):
    red[i] = i
    green[i] = i
    mi[i] = 100 * np.log10(1/(i + 1))
    mi_norm = (mi * 0.0042 + 1) * 255
```

```python
fig, ax = plt.subplots(layout='constrained', figsize=(10, 6), nrows=1, ncols=2)
ax[0].scatter(red, mi_norm)
ax[0].set_ylabel('MI normalized')
ax[0].set_xlabel('Red channel')

ax[1].scatter(red, mi)
ax[1].set_xlabel('Red channel')
ax[1].set_ylabel('MI')
fig.suptitle('Red channel transformation into MI signal')
plt.show()
```

```python
print(f'Min of MI {mi.min():.02f}')
print(f'Max of MI {mi.max():0.2f}\n')

print(f'Min of normalized MI {mi_norm.min():.02f}')
print(f'Max of normalized MI {mi_norm.max():.02f}')
```

```python
red = np.arange(0, 255, 1)
green = np.arange(0, 255, 1)
red, green = np.meshgrid(red, green)
EI = 100 * (np.log10(1/(green.astype('float')+1)) - 1.44 * np.log10(1/(red.astype('float')+1)))
EI_norm = (EI * 0.0017 + 0.4098) * 255
```

```python
fig = plt.figure(figsize = (16, 8))

ax = fig.add_subplot(1, 2, 1, projection='3d')
surf = ax.plot_surface(red, green, EI_norm, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter('{x:.0f}')
ax.set_xlabel('Red channel')
ax.set_ylabel('Green channel')
ax.set_zlabel('EI norm')
# fig.colorbar(surf, shrink=0.5)

ax = fig.add_subplot(1, 2, 2, projection='3d')
surf = ax.plot_surface(green, red, EI_norm, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter('{x:.0f}')
ax.set_xlabel('Green channel')
ax.set_ylabel('Red channel')
ax.set_zlabel('EI norm')

# fig.colorbar(surf, shrink=0.5, aspect=5)
fig.suptitle('Red and Green channels translation into EI signal')
plt.show()

```

```python
print(f'Min of EI {EI.min():.02f}')
print(f'Max of EI {EI.max():0.2f}\n')

print(f'Min of normalized EI {EI_norm.min():.02f}')
print(f'Max of normalized EI {EI_norm.max():.02f}')
```
